```python
def mergeColumns(data):
    data['Date'] = pd.to_datetime(data['Date'])
    data['BTX'] = data['Benzene'] + data['Toluene'
      ] + data['Eth-Benzene']
    data.drop(['Benzene', 'Toluene', 'Eth-Benzene'], axis=1)
    data['Particulate_Matter'] = data['PM2.5 '] + data['PM10 ']
    return data

def subsetColumns(data):
    pollutants = ['Particulate_Matter', 'NO2 ', 'CO ','SO2 ', 'Ozone ', 'BTX']
    columns =  ['Date', 'City', 'AQI', 'AQI Bucket'] + pollutants
    data = data[columns]
    return data, pollutants
def handleMissingValues(data):
   # missing_values = getMissingValues(data)
    newAQI = mergeColumns(data)
    newAQI, pollutants = subsetColumns(newAQI)
    return newAQI, pollutants

newAQI, newColumns = handleMissingValues(AQI)
newAQI
```

```python
#normalization of data
min_Particulate_Matter = newAQI['Particulate_Matter'].min()
max_Particulate_Matter = newAQI['Particulate_Matter'].max()
newAQI['Particulate_Matter_new'] = [(x-min_Particulate_Matter)/(max_Particulate_Matter-min_Particulate_Matter) 
                                         for x in newAQI['Particulate_Matter']]

min_NO2	= newAQI['NO2 '].min()
max_NO2 = newAQI['NO2 '].max()
newAQI['NO2_new'] = [(x-min_NO2)/(max_NO2-min_NO2) for x in newAQI['NO2 ']]

min_CO= newAQI['CO '].min()
max_CO = newAQI['CO '].max()
newAQI['CO_new'] = [(x-min_CO)/(max_CO-min_CO) for x in newAQI['CO ']]

min_SO2= newAQI['SO2 '].min()
max_SO2 = newAQI['SO2 '].max()
newAQI['SO2_new'] = [(x-min_SO2)/(max_SO2-min_SO2) for x in newAQI['SO2 ']]

min_O3= newAQI['Ozone '].min()
max_O3 = newAQI['Ozone '].max()
newAQI['Ozone _new'] = [(x-min_O3)/(max_O3-min_O3) for x in newAQI['Ozone ']]

min_BTX = newAQI['BTX'].min()
max_BTX = newAQI['BTX'].max()
newAQI['BTX_new'] = [(x-min_BTX)/(max_BTX-min_BTX) for x in newAQI['BTX']]
newAQI
```

```python
def visualisepollutants(udata, column):
    data = udata.copy()
    data.set_index('Date',inplace=True)
            
    axes = data[column].plot(marker='.', alpha=10, linestyle='None', figsize=(16, 15), subplots=True)
    for ax in axes:
        ax.set_xlabel('Years')
        ax.set_ylabel('ug/m3')
visualisepollutants(nAQI, pollutant)
```

```python
def trend_plot(nAQI,value):
    data=nAQI.copy()
    data['Year']=[d.year for d in data.Date]
    data['Month'] = [d.strftime('%b') for d in data.Date]
    years= data['Year'].unique()
    fig, axes = plt.subplots(1, 2, figsize=(12,3), dpi= 80)
    sns.boxplot(x='Year', y=value, data=data, ax=axes[0])
    sns.lineplot(x='Month', y=value, data=data.loc[~data.Year.isin([2019, 2021]), :])
    
    axes[0].set_title('Year-wise Plot i.e. the trend', fontsize=18); 
    axes[1].set_title('Month-wise Plot i.e. the seasonality', fontsize=18)
    plt.show()
value='Particulate_Matter_new'
trend_plot(nAQI,value)
value='NO2_new'
trend_plot(nAQI,value)
value='CO_new'
trend_plot(nAQI,value)
value='SO2_new'
trend_plot(nAQI,value)
value='Ozone _new'
```

```python
AQI_pivot = nAQI.pivot(index='Date',columns='City', values='AQI')
AQI_pivot.head()
```

```python
def get_colour_bar(City):
    col=[]
    for val in AQI_pivot[City]:
        if val<50:
            col.append("Green")
        elif val>50  and val<101:
            col.append("Yellow")
        elif val>100 and val<151:
            col.append("Orange")
        elif val>150 and val<201:
            col.append("Red")
        elif val>200 and val<301:
            col.append("purple")
        else:
            col.append("maroon")
    return col

bp = get_colour_bar('Bhopal')
mn = get_colour_bar('Mandideep')
colors = {'Good':'Green','Satisfactory':'Yellow','Moderate':'Orange','Poor':'Red','Very Poor':'purple','Severe':'maroon'}
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
f, ((ax1, ax2)) = plt.subplots(2, 1, sharex='col', sharey='row', figsize=(20,10))
ax1.bar(AQI_pivot.index, AQI_pivot['Bhopal'], color = bp, width = 0.75)
ax2.bar(AQI_pivot.index, AQI_pivot['Mandideep'], color = mn, width = 0.75)
ax1.legend(handles, labels, loc='upper left')
ax2.legend(handles, labels, loc='upper left')
ax1.title.set_text('Bhopal')
ax1.title.set_text('Mandideep')
ax1.set_ylabel('AQI')
ax2.set_ylabel('AQI')

        
```

```python
cities=['Bhopal','Mandideep']
AQI_beforeLockdown = AQI_pivot['2019-01-01':'2020-03-25']
AQI_duringLockdown = AQI_pivot['2020-03-26':'2020-07-01']
AQI_afterLockdown =  AQI_pivot['2020-07-02':'2021-12-31']
limits = [50, 100, 200, 300, 400, 510]
#palette = sns.light_palette("Spectral", len(limits), reverse = True)
palette = sns.color_palette("coolwarm", len(limits))

for city in cities:
    aqi_before = AQI_beforeLockdown[city].mean()
    aqi_during = AQI_duringLockdown[city].mean()
    aqi_after  = AQI_afterLockdown[city].mean()
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=(27, 1.5))
    ax1.set_yticks([1])
    ax1.set_yticklabels([city])
    ax1.spines['bottom'].set_visible(False)
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax1.spines['left'].set_visible(False)

    prev_limit = 0
    for idx, lim in enumerate(limits):
        ax1.barh([1], lim-prev_limit, left=prev_limit, height=15, color=palette[idx])
        prev_limit = lim

    ax1.barh([1], aqi_before, color='black', height=5) 
    #during lockdown
    ax2.set_yticks([1])
    ax2.set_yticklabels([city])
    ax2.spines['bottom'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['left'].set_visible(False)

    prev_limit = 0
    for idx, lim in enumerate(limits):
        ax2.barh([1], lim-prev_limit, left=prev_limit, height=15, color=palette[idx])
        prev_limit = lim

    ax2.barh([1], aqi_after, color='black', height=5)
    
   
    ax1.set_title('Before Lockdown')
    ax2.set_title('During Lockdown')
    
    
    rects = ax1.patches
    labels=["Good", "Satisfactory", "Moderate", "Poor", 'Very Poor', 'Severe']
    
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax1.text(
            rect.get_x() + rect.get_width() / 2,
            -height * .4,
            label,
            ha='center',
            va='bottom',
            color='black')
        ax2.text(            rect.get_x() + rect.get_width() / 2,
            -height * .4,
            label,
            ha='center',
            va='bottom',
            color='black')

 
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(timeseries):
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
       dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)

print(adf_test(Bh_Seasonal_AQI))
```

```python
#adjusting the AQI
Adj_Bhopal_AQI= Bh_Seasonal_AQI-Bh_Seasonal_AQI.shift(12)
Adj_Bhopal_AQI=Adj_Bhopal_AQI.dropna()
Adj_Bhopal_AQI.plot()
print(adf_test(Adj_Bhopal_AQI))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima;
```

```python
auto_arima(y=Bh_Seasonal_AQI,start_p=0,start_P=0,start_q=0,start_Q=0,seasonal=True, m=6)
```

```python
train = Bh_Seasonal_AQI[:24] #from 2019-2020
test = Bh_Seasonal_AQI[24:36]# 2021
```

```python
test
```

```python
model=SARIMAX(train,order=(1,0,0),seasonal_order=(1,0,1,12),)
results=model.fit()
results.summary()
```

```python
predictions = results.predict(start=24, end=35, typ='levels').rename('Predictions')
```

```python
predictions.plot(legend=True)
test.plot(legend=True,title="Bhopal Prediction data");
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score, max_error,mean_poisson_deviance,mean_gamma_deviance
import math
RMSE=np.sqrt(mean_squared_error(predictions,test))
print('Root Mean Squared Error: ', RMSE)
print('Mean AQI:',test.mean())
forecast_errors = [test[i]-predictions[i] for i in range(len(test))]
bias = sum(forecast_errors) * 1.0/len(test)
print('Bias: %f' % bias)
mse = mean_squared_error(test, predictions)
print('MSE: '+str(mse))
mae = mean_absolute_error(test, predictions)
print('MAE: '+str(mae))
rmse = math.sqrt(mean_squared_error(test, predictions))
print('RMSE: '+str(rmse))
mape = np.mean(np.abs(predictions - test)/np.abs(test))
print('MAPE: '+str(mape))
r2score=r2_score(test, predictions)
print('r2score: '+str(r2score))
explainedVariance_score=explained_variance_score(test, predictions)
print('explainedVariance_score: '+str(explainedVariance_score))
me=max_error(test, predictions)
print('me: '+str(me))
mpd=mean_poisson_deviance(test, predictions)
print('mpd: '+str(mpd))
mgd=mean_gamma_deviance(test, predictions)
print('mgd: '+str(mgd))
```

```python
final_model = SARIMAX(train,order=(1,0,0),seasonal_order=(1,0,1,12))
results = final_model.fit()
#Obtaining predicted values:
predictions = results.predict(start=36, end=48, typ='levels').rename('Predictions')
#Plotting predicted values against the true values:
predictions.plot(legend=True)
Bh_Seasonal_AQI.plot(legend=True,figsize=(12,8),grid=True,title="Bhopal AQI");
```

```python
df_anova = pd.read_csv('newAQI.csv')
df_anova = df_anova[['AQI','City']]
```

```python
from scipy import stats
Citys = pd.unique(df_anova.City.values)
d_data = {city:df_anova['AQI'][df_anova.City == city] for city in Citys}
 
F, p = stats.f_oneway(d_data['Mandideep'], d_data['Bhopal'])

print("p-value for significance is: ", p)

if p<0.05:
    print("We reject the null hypothesis")
else:
    print("We accept the null hypothesis")
```
